```python
import pandas as pd
import numpy as np
```

```python
def isfloat(num):
    try:
        float(num)
        return True
    except ValueError:
        return False
```

```python
def remove_items(test_list, item):

    # using list comprehension to perform the task
    res = [i for i in test_list if i != item]
    return res
```

```python
def unique(list1):

    # initialize a null list
    unique_list = []

    # traverse for all elements
    for x in list1:
        # check if exists in unique_list or not
        if x not in unique_list:
            unique_list.append(x)
    return(unique_list)
```

```python
df = pd.read_csv("/content/Copy of LawYantra Dataset - lawyer_summaries.csv")
```

```python
exp=[]
prac=[]
prac2=[]
feed=[]
feed2=[]
jur=[]
rate=[]
rate2=[]
dispT=[]
dispT2=[]
lang=[]
city=[]
comp=[]
demo=[]
prob=[]

for index, row in df.iterrows():
  split_str1 = row["Information"].split(". ")

  p1=split_str1[0].split(" ")

  exp.append(p1[p1.index("has")+1])

  prac.append(p1[p1.index("in")+1:])

  p2=split_str1[1].split(" ")

  feed.append(p2[p2.index("of")+1])

  p3=split_str1[2].split(" ")

  jur.append(" ".join(p3[p3.index("is")+1:]))

  p4=split_str1[3].split(" ")

  rate.append(p4[p4.index("charges")+1])

  p5=split_str1[4].split(" ")

  dispT.append(p5[p5.index("takes")+1])

  p6=split_str1[5].split(" ")

  lang.append(p6[p6.index("speaks:")+1:])

  p7=split_str1[6].split(" ")

  city.append(p7[-1])

  temp = p7[p7.index("at")+1:p7.index("is")-1]
  temp=" ".join(temp)
  comp.append(temp[:-1])

  p8=split_str1[7].split(" ")

  if "not" in p8:
    prob.append(0)
  else:
    prob.append(1)

  p9=split_str1[8].split(" ")

  demo.append(" ".join(p9[p9.index("is")+1:])[:-1])



for i in feed:
  if isfloat(i):
    feed2.append(eval(i))
  else:
    feed2.append(None)

for i in rate:
  if isfloat(i):
    rate2.append(eval(i))
  else:
    rate2.append(None)

for i in dispT:
  if isfloat(i):
    dispT2.append(eval(i))
  else:
    dispT2.append(None)

for i in lang:
  i.pop((i.index('and')))
  for x in i:
    if x[-1] == ',':
      lang[lang.index(i)][i.index(x)]=x[:-1]

for i in prac:
  prac2.append(remove_items(i, "Law,"))



exp = [eval(i) for i in exp]
```

```python
df["exp"]=exp
df["prac"]=prac2
df["feed"]=feed2
df["jur"]=jur
df["rate"]=rate2
df["dispT"]=dispT2
df["lang"]=lang
df["city"]=city
df["comp"]=comp
df["demo"]=demo
df["prob"]=prob
df.to_csv("Lawyer Info Cleaned.csv")
```

```python
law = pd.read_csv("/content/law_words.csv")
```

```python
law.head()
```

```python
from nltk .stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y=[]

  for i in text.split():
   y.append(ps.stem(i))
  return " ".join(y)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
cv= CountVectorizer(max_features=5000,stop_words='english')
def recommendCV(query):
  temp = pd.concat([law["words"], pd.Series(query)])
  temp = temp.apply(stem)
  vectors=cv.fit_transform(temp).toarray()
  similarity=cosine_similarity(vectors)
  law_index=-1
  diatance= similarity[law_index]
  names_list=sorted(list(enumerate(similarity[law_index])),reverse=True,key=lambda x:x[1])[1:6]

  for i in names_list:
    print(law.iloc[i[0]]["cat"], round(i[1]*100))
```
